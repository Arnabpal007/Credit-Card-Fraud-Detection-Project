# Credit Card Fraud Detection Project

In this project we will be working with a creditcard data set, indicating whether a perticular transaction in our collected data is Legit or Fraud.
We will try to create a model that will predict whether the transaction is Legit or Fraud.

In our creditcard data set:
* '0': Legit Transaction
* '1': Fraud Transaction

## Import Libraries

**Importing required libraries**

```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline
```

## Get the Data
**Reading the creditcard.csv file and set it to a data frame called creditcard_data.**

```python
creditcard_data=pd.read_csv('creditcard.csv')
```

**Checking the head of creditcard_data.**

```python
creditcard_data.head()
```

**Using info() and describe() on creditcard_data.**

```python
creditcard_data.info()
```

```python
creditcard_data.describe()
```

## Exploratory Data Analysis

**Let's explore the data.**

**Checking number of rows and columns.**

```python
creditcard_data.shape
```

**Creating a countplot showing how many of the transactions are Legit or Fraud in the data.**

```python
fig=plt.figure(figsize=(10,6))
fig.patch.set_facecolor('xkcd:mint green')
cp=sns.countplot(x='Class',data=creditcard_data)
for p in cp.patches:
   cp.annotate('{:}'.format(p.get_height()), (p.get_x()+0.36, p.get_height()+0.01))
```

**We can see that the data set is highly unbalanced.**

**Separating the data for more analysis.**

```python
legit_data=creditcard_data[creditcard_data['Class']==0]
fraud_data=creditcard_data[creditcard_data['Class']==1]
```

```python
print(legit_data.shape)
print(fraud_data.shape)
```

**Analysing the legit_data and fraud_data statically.**

```python
legit_data.describe()
```

```python
fraud_data.describe()
```

**Comparing the mean of each columns of the creditcard_data  based on wheather the transaction is Legit or Fraud using groupby.** 

```python
creditcard_data.groupby('Class').mean()
```

## Under-Sampling
**Building a sample data set containg the similar distribution of Legit and Fraud Transactions.** 

**Creating a sample of legit_data.**

```python
sample_legit_data=legit_data.sample(n=492)
```

```python
sample_legit_data.head()
```

```python
sample_legit_data.shape
```

**Concatenating sample_legit_data and fraud_data to create a new well-balanced creditcard data.**

```python
new_creditcard_data=pd.concat((sample_legit_data,fraud_data),axis=0)
```

```python
new_creditcard_data.head()
```

```python
new_creditcard_data['Class'].value_counts()
```

**Now we can make our ML model based on this new well-balanced data set.**

## Logistic Regression Model
Now we have to create a Logistic Regression Model for our data set.

**Split the data into training set and testing set using train_test_split.**

```python
from sklearn.model_selection import train_test_split
```

```python
X=new_creditcard_data.drop(columns='Class',axis=1)
y=new_creditcard_data['Class']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=101)
```

**Training and fitting the logistic regression model on the training set.**

```python
from sklearn.linear_model import LogisticRegression
```

```python
detectFraud_model=LogisticRegression()
detectFraud_model.fit(X_train,y_train)
```

## Predictions and Evaluations

**Predicting the values for the testing data.**

```python
predictions=detectFraud_model.predict(X_test)
```

**Creating a confusion matrix and classification report to see how well the model works.**

```python
from sklearn.metrics import confusion_matrix,classification_report
```

```python
print(confusion_matrix(y_test,predictions))
```

```python
print(classification_report(y_test,predictions))
```

**According to the confusion matrix and classification report our model is working quite well.**
